```python
import numpy as np
```

```python
#creating two dimensional arrays
n1 = np.array([[1,2,3],[4,5,6]])
print(n1)
```

```python
#array filled with zeros
n3 = np.zeros([2,4])
print(n3)
```

```python
#array filled with ones
n4 = np.ones([2,4])
print(n4)
```

```python
#two array with 3 by 3 matrix
n2 = np.ones([2,3,3])
print(n2)
```

```python
#empty array with random values
n5 = np.empty([2,3])
print(n5)
```
